```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

```

```python
import pandas as pd
```

```python
stats = pd.read_csv(r'P4-Demographic-Data.csv')
```

```python
stats.head()
```

```python
plt.rcParams['figure.figsize'] = 8,4
```

```python
#Distribution

vis1 = sns.distplot(stats["Internet users"])
```

```python
vis2 = sns.distplot(stats["Birth rate"])
```

```python
vis3 = sns.distplot(stats["Internet users"],bins=30)
```

```python
vis3 = sns.boxplot(data=stats,x="Income Group", y = "Birth rate")
```

```python
#linear model plot
vis4 = sns.lmplot (data = stats, x = 'Internet users', y= 'Birth rate', fit_reg = False)
```

```python
#breaking the same plot with internet users
vis5 = sns.lmplot (data = stats, x = 'Internet users', y= 'Birth rate', 
                   fit_reg = False, hue='Income Group',size=8,
                  scatter_kws={"s":200})
```

From the plotted lmplot graph it is interesting to note the following:
    1. Internet Users/ People with high income have less birth rate
    2. internet Users/ People with low income have high birth rate
    As the number of internet users increase, birth rate shows a fall as indicated by graph.
