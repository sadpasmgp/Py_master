Since there are 36 Studios who produce movies and 15 Genres, we will try to filter to 6 studios and 5 Genres

```python
mov2 = mov[(mov.Studio == 'Fox')|(mov.Studio == 'WB')|(mov.Studio == 'Sony')
           |(mov.Studio == 'Lionsgate')|(mov.Studio == 'Universal')|(mov.Studio == 'MGM')]
mov2
```

As you can see to get 6 studios we will have to right lots of code instead we use a different approach

```python
studio_filters = ['Fox','WB','Sony','Lionsgate','Universal','MGM']
mov3 = mov[mov.Studio.isin(studio_filters)]

genre_filters = ['action','adventure','animation','comedy','drama']
mov3 = mov[mov.Genre.isin(genre_filters)]
```

```python
len(mov3)
```

```python
sns.set(style='darkgrid',palette = 'muted',color_codes = True)

ax = sns.boxplot(data = mov3, x='Genre', y='Gross % US', orient = 'v')
plt.setp(ax.artists, alpha = 0.5)

sns.stripplot(x= 'Genre', y = 'Gross % US', data = mov3, jitter = True, 
              size = 6, linewidth = 0, hue = 'Studio', alpha = 0.7)

ax.axes.set_title('Domestic Gross % by Genre', fontsize = 30)
ax.set_xlabel('Genre',fontsize = 20)
ax.set_ylabel('Gross % US', fontsize = 20)
ax.legend(bbox_to_anchor=(1,1))
```
